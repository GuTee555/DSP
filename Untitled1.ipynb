```python
#ประกาศเพื่อเรียกใช้งาน
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

```python
#lab01
sin_wave=thinkdsp.SinSignal(freq=7019,amp=19,offset=0)
sin_wave.plot()
thinkplot.config(xlabel='time (s)')
```

```python
#Lab02
#Trianglesignal(f)
#duration=signal.period*n

signal =thinkdsp.SinSignal(4000)
duration =  signal.period*2500
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
teewave=signal.make_wave(duration, framerate=10000)
teewave.apodize
teewave.make_audio()
```

```python
signal=thinkdsp.SquareSignal(2000)
duration =  signal.period*200
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
thinkplot.config(ylin=[-1.05,1.05],legend=False)
segment.plot()
sqwave=signal.make_wave(duration, framerate=10000)
sqwave.apodize()
sqwave.make_audio()
```
